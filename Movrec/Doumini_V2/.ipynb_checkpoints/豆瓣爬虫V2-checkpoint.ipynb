```python
# ==== ç¬¬ä¸€éƒ¨åˆ†ï¼šå¯¼å…¥å’ŒåŸºç¡€è®¾ç½® ====
import requests
import urllib3
import ssl
import pandas as pd
import time
import random
from requests.adapters import HTTPAdapter
from typing import Optional
from urllib3.util.retry import Retry

# å½»åº•ç¦ç”¨æ‰€æœ‰ SSL è­¦å‘Šå’ŒéªŒè¯
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# ==== ç¬¬äºŒéƒ¨åˆ†ï¼šSSL ç›¸å…³ç±»å’Œå‡½æ•° ====
class InsecureHTTPAdapter(HTTPAdapter):
    """å®Œå…¨ç¦ç”¨ SSL éªŒè¯çš„è‡ªå®šä¹‰é€‚é…å™¨"""
    
    def init_poolmanager(self, *args, **kwargs):
        kwargs['ssl_context'] = self._create_ssl_context()
        return super().init_poolmanager(*args, **kwargs)
    
    def proxy_manager_for(self, *args, **kwargs):
        kwargs['ssl_context'] = self._create_ssl_context()
        return super().proxy_manager_for(*args, **kwargs)
    
    def _create_ssl_context(self):
        context = ssl.create_default_context()
        context.check_hostname = False
        context.verify_mode = ssl.CERT_NONE
        return context

def create_insecure_session():
    """åˆ›å»ºå®Œå…¨ç¦ç”¨ SSL éªŒè¯çš„ä¼šè¯"""
    session = requests.Session()
    insecure_adapter = InsecureHTTPAdapter()
    session.mount('https://', insecure_adapter)
    session.mount('http://', insecure_adapter)
    session.verify = False
    return session

# ==== ç¬¬ä¸‰éƒ¨åˆ†ï¼šå…¨å±€å˜é‡å’Œä¼šè¯åˆå§‹åŒ– ====
# ç”¨æˆ·ID
USER_ID = "253231903"

# User-Agent
MOBILE_UA = "Mozilla/5.0 (iPhone; CPU iPhone OS 14_0 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/14.0 Mobile/15E148 Safari/604.1"

# åˆ›å»ºå…¨å±€ä¼šè¯
sess = create_insecure_session()
sess.headers.update({
    'User-Agent': MOBILE_UA,
    'Accept': 'application/json, text/plain, */*',
    'Accept-Language': 'zh-CN,zh;q=0.9,en;q=0.8',
})

# ==== ç¬¬å››éƒ¨åˆ†ï¼šæ ¸å¿ƒè¯·æ±‚å‡½æ•° ====
def robust_get(sess, url, headers, timeout=10, **kwargs):
    """ä¿®å¤ç‰ˆçš„è¯·æ±‚å‡½æ•° - ç›´æ¥ä½¿ç”¨ä¸å®‰å…¨ä¼šè¯"""
    SLEEP_BASE = 1
    jitter = random.uniform(0.5, 1.5)
    
    try:
        print("ä½¿ç”¨ä¸å®‰å…¨ä¼šè¯è¿›è¡Œè¯·æ±‚...")
        insecure_sess = create_insecure_session()
        insecure_sess.headers.update(sess.headers)
        insecure_sess.cookies.update(sess.cookies)
        
        r = insecure_sess.get(url, headers=headers, timeout=timeout, **kwargs)
        time.sleep(SLEEP_BASE + jitter)
        
        if r.status_code == 200:
            print("è¯·æ±‚æˆåŠŸï¼")
            return r
        else:
            print(f"è¯·æ±‚è¿”å›çŠ¶æ€ç : {r.status_code}")
            r.raise_for_status()
            
    except Exception as e:
        print(f"è¯·æ±‚å¤±è´¥: {e}")
        try:
            print("ä½¿ç”¨æœ€ç®€åŒ–è¯·æ±‚...")
            response = requests.get(url, headers=headers, timeout=timeout, verify=False, **kwargs)
            time.sleep(SLEEP_BASE + jitter)
            return response
        except Exception as final_error:
            print(f"æœ€ç»ˆå°è¯•ä¹Ÿå¤±è´¥: {final_error}")
            raise

# ==== ç¬¬äº”éƒ¨åˆ†ï¼šæ•°æ®è·å–å‡½æ•° ====
def api_fetch_user_interests(user_id, status, count=15, max_pages=100):
    """è·å–ç”¨æˆ·ç”µå½±å…´è¶£åˆ—è¡¨"""
    api = f"https://m.douban.com/rexxar/api/v2/user/{user_id}/interests"

    params = {
        "type": "movie", 
        "status": status,  # ä¿æŒå‚æ•°ä¸å˜
        "count": count,
        "start": 0,
        "order_by": "time",
    }

    headers = {
        "User-Agent": MOBILE_UA,
        "Accept": "application/json, text/plain, */*",
        "Origin": "https://m.douban.com",
        "Referer": f"https://m.douban.com/movie/people/{user_id}/",
        "X-Requested-With": "XMLHttpRequest",
    }

    print(f"å¼€å§‹è·å– {status} æ•°æ®...")
    
    all_items = []
    page = 0

    while page < max_pages:
        print(f"è¯·æ±‚ç¬¬{page+1}é¡µï¼Œstart={params['start']}...")
        r = robust_get(sess, api, headers=headers, params=params, timeout=15)
        data = r.json()
        
        items = data.get("interests") or data.get("items") or []
        print(f"è·å–åˆ° {len(items)} æ¡æ•°æ®")
        
        if not items:
            break
            
        all_items.extend(items)
        
        if len(items) < count:
            break
            
        params["start"] += count
        page += 1
        time.sleep(1.2)

    # å»é‡
    seen = set()
    deduped = []
    for it in all_items:
        sid = str(((it or {}).get('subject') or {}).get('id') or '')
        key = f"{sid}"
        if sid and key not in seen:
            seen.add(key)
            deduped.append(it)

    print(f"æ€»å…±è·å– {len(deduped)} æ¡ç”µå½±æ•°æ®")
    return deduped

# ==== ç¬¬å…­éƒ¨åˆ†ï¼šæ•°æ®å¤„ç†å‡½æ•° ====
import re  # ğŸ”¥ æ·»åŠ è¿™è¡Œå¯¼å…¥

# ğŸ”¥ æ·»åŠ ç¼ºå¤±çš„å˜é‡å®šä¹‰
_CJK_RGX = re.compile(r'[\u4e00-\u9fff]')
_ASCII_RATIO_TH = 0.6  # è‹±æ–‡ååˆ¤æ–­é˜ˆå€¼ï¼šASCII æ¯”ä¾‹ > 60% è®¤ä¸ºæ›´åƒè‹±æ–‡

def _looks_non_cjk(s: str) -> bool:
    if not s:
        return False
    # å«æœ‰æ±‰å­— => è®¤ä¸ºä¸æ˜¯è‹±æ–‡
    if _CJK_RGX.search(s):
        return False
    # è¿›ä¸€æ­¥ç”¨ ASCII å æ¯”åšä¸ªç®€å•é²æ£’åˆ¤æ–­
    ascii_cnt = sum(1 for ch in s if ord(ch) < 128)
    return (ascii_cnt / max(1, len(s))) >= _ASCII_RATIO_TH

def pick_english_title(subject: dict) -> str:
    """ä» subject ä¸­å°½å¯èƒ½æŒ‘ä¸€ä¸ªåƒ"è‹±æ–‡å"çš„æ ‡é¢˜"""
    # 1) original_titleï¼ˆå¦‚æœåŸºæœ¬éä¸­æ–‡ï¼‰
    ot = (subject or {}).get('original_title') or ''
    if _looks_non_cjk(ot):
        return ot.strip()

    # 2) aka åˆ«ååˆ—è¡¨é‡ŒæŒ‘ä¸€ä¸ªæ›´åƒè‹±æ–‡çš„
    aka_list = (subject or {}).get('aka') or []
    for name in aka_list:
        if _looks_non_cjk(name):
            return name.strip()

    # 3) å…œåº•ï¼šå¦‚æœ title çœ‹èµ·æ¥å°±å¾ˆåƒè‹±æ–‡ï¼Œä¹Ÿè¿”å›ï¼›å¦åˆ™ç©º
    title = (subject or {}).get('title') or ''
    if _looks_non_cjk(title):
        return title.strip()
    return ''

def crawl_my_movies_via_mobile_api(user_id: str, status: str, limit: Optional[int]=None, comments_n: int=5) -> pd.DataFrame:
    """
    status: 'collect' çœ‹è¿‡ | 'wish' æƒ³çœ‹
    """
    rows = api_fetch_user_interests(user_id, status)
    
    # çŠ¶æ€è¿‡æ»¤
    if status == "collect":
        filtered_rows = [item for item in rows if item.get('status') in ['mark', 'done']]
    elif status == "wish":
        filtered_rows = [item for item in rows if item.get('status') == 'wish']
    
    print(f"çŠ¶æ€è¿‡æ»¤: {len(rows)} -> {len(filtered_rows)} æ¡{status}æ•°æ®")
    
    if limit:
        filtered_rows = filtered_rows[:limit]

    if not filtered_rows:
        print("âŒ è¿‡æ»¤åæ²¡æœ‰æ•°æ®")
        return pd.DataFrame()

    out = []
    error_count = 0

    for i, item in enumerate(filtered_rows):
        try:
            print(f"å¤„ç†ç¬¬ {i+1}/{len(filtered_rows)} æ¡æ•°æ®...")
            
            subject = item.get('subject') or {}
            if not subject:
                print(f"  âš ï¸ ç¬¬{i+1}æ¡: subjectä¸ºç©º")
                error_count += 1
                continue
                
            title = subject.get('title', '').strip()
            if not title:
                print(f"  âš ï¸ ç¬¬{i+1}æ¡: æ ‡é¢˜ä¸ºç©º")
                error_count += 1
                continue

            print(f"  âœ… å¤„ç†: {title}")

            # è‹±æ–‡å
            english_name = pick_english_title(subject)

            # è¯­è¨€ä¿¡æ¯
            languages = subject.get('languages') or []
            language_str = ', '.join([x for x in languages if x]) if languages else ''

            # æˆ‘çš„è¯„åˆ†
            my_rating = 0
            r_obj = item.get('rating')
            if isinstance(r_obj, dict):
                my_rating = r_obj.get('value', 0) or 0

            # è±†ç“£å…¬å…±è¯„åˆ†
            douban_rating = 0
            rating_count = 0
            sr = subject.get('rating') or {}
            if isinstance(sr, dict):
                douban_rating = sr.get('value', 0) or 0
                rating_count = sr.get('count', 0) or 0

            # ç±»å‹
            genres = subject.get('genres') or []
            genres_str = ', '.join(genres) if genres else ''

            # å…¶ä»–å­—æ®µ
            alias_str = ', '.join(subject.get('aka', []) or [])
            release_date = subject.get('release_date', '') or ''
            short_comment = item.get('comment', '') or ''
            mark_time = item.get('create_time', '') or ''
            douban_id = str(subject.get('id', '') or '')

            movie_data = {
                'name': title,
                'english_name': english_name,
                'language': language_str,
                'alias': alias_str,
                'vote_average': douban_rating,
                'vote_count': rating_count,
                'release_date': release_date,
                'my_rating': my_rating,
                'short_comments': short_comment,
                'mark_time': mark_time,
                'douban_id': douban_id,
                'genres': genres,
                'genres_str': genres_str,
            }
            out.append(movie_data)

        except Exception as e:
            print(f"  âŒ ç¬¬{i+1}æ¡å¤„ç†å¼‚å¸¸: {e}")
            error_count += 1
            continue

    print(f"å¤„ç†ç»“æœ: æˆåŠŸ {len(out)} æ¡, å¤±è´¥ {error_count} æ¡")
    return pd.DataFrame(out)

# ==== ç¬¬ä¸ƒéƒ¨åˆ†ï¼šæ‰§è¡Œä»£ç  ====
if __name__ == "__main__":
    print("=" * 50)
    print("å¼€å§‹æ‰§è¡Œè±†ç“£æ•°æ®è·å–...")
    print("=" * 50)
    
    # ====== ç¤ºä¾‹ï¼šå…ˆå„å– 20 æ¡è¯•è·‘ ======
    print("\n1. è·å–æ”¶è—æ•°æ®...")
    df_collect_optimized = crawl_my_movies_via_mobile_api(USER_ID, "collect", limit=20, comments_n=5)
    
    print(f"\næ”¶è—æ•°æ®è·å–å®Œæˆï¼Œå…± {len(df_collect_optimized)} æ¡")
    
    print("\n2. è·å–æƒ³çœ‹æ•°æ®...")
    df_wish_optimized = crawl_my_movies_via_mobile_api(USER_ID, "wish", limit=20, comments_n=5)
    
    print(f"\næƒ³çœ‹æ•°æ®è·å–å®Œæˆï¼Œå…± {len(df_wish_optimized)} æ¡")
    
    # æŸ¥çœ‹ä¼˜åŒ–åçš„å­—æ®µ
    print("\n" + "=" * 50)
    print("æ•°æ®é¢„è§ˆ:")
    print("=" * 50)
    
    print("\nä¼˜åŒ–åçš„æ•°æ®å­—æ®µ:")
    print(df_collect_optimized.columns.tolist())

    print("\næ”¶è—æ•°æ®æ ·ä¾‹:")
    display(df_collect_optimized[['name', 'english_name', 'language', 'my_rating', 'genres_str']].head(10))

    print("\nè¯­è¨€åˆ†å¸ƒ:")
    print(df_collect_optimized['language'].value_counts())
    
    print("\nç±»å‹åˆ†å¸ƒ:")
    print(df_collect_optimized['genres_str'].value_counts().head(10))
    
    print("=" * 50)
    print("æ‰§è¡Œå®Œæˆ!")
    print("=" * 50)
# ==== ç¬¬å…«éƒ¨åˆ†ï¼šæ•°æ®ä¿å­˜ ====
    print("\n" + "=" * 50)
    print("æ•°æ®ä¿å­˜...")
    print("=" * 50)
    
    # ä¿å­˜æ”¶è—æ•°æ®
    df_collect_optimized.to_csv('æˆ‘çš„è±†ç“£æ”¶è—ç”µå½±.csv', index=False, encoding='utf-8-sig')
    print(f"âœ… æ”¶è—æ•°æ®å·²ä¿å­˜: æˆ‘çš„è±†ç“£æ”¶è—ç”µå½±.csv ({len(df_collect_optimized)} æ¡)")
    
    # ä¿å­˜æƒ³çœ‹æ•°æ®ï¼ˆå¦‚æœæœ‰çš„è¯ï¼‰
    if len(df_wish_optimized) > 0:
        df_wish_optimized.to_csv('æˆ‘çš„è±†ç“£æƒ³çœ‹ç”µå½±.csv', index=False, encoding='utf-8-sig')
        print(f"âœ… æƒ³çœ‹æ•°æ®å·²ä¿å­˜: æˆ‘çš„è±†ç“£æƒ³çœ‹ç”µå½±.csv ({len(df_wish_optimized)} æ¡)")
    
    print("=" * 50)
    print("æ‰§è¡Œå®Œæˆ!")
    print("=" * 50)
```
